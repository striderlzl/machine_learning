# ii. Test all the data in the test database with k nearest neighbors. Take decisions by majority polling. Plot train and test errors in terms of k for k  {208, 205, . . . , 7, 4, 1, } (in reverese order). You are welcome to use smaller increments of k. Which k is the most suitable k among those values? Cal- culate the confusion matrix, true positive rate, true negative rate, precision, and F -score when k = k

```python
#k from 208 to 1
def find_ks(train_x,train_y,test_x,test_y,start,end,step):
    error =[]
    testing_scores = []
    training_scores = []
    k=list(range(start,end,step))
    for i in range(start,end,step):
        cl_ = KNeighborsClassifier(n_neighbors=i)  
        cl_.fit(train_x,train_y)
        training_scores.append(cl_.score(train_x,train_y))
        testing_scores.append(cl_.score(test_x,test_y))
        pred_i=cl_.predict(test_x)
        error.append(np.mean(pred_i - test_y))
        opt_k=k[testing_scores.index(max(testing_scores))]
    return(training_scores,testing_scores,opt_k)
training_scores,testing_scores,opt_k = find_ks(train_x,train_y,test_x,test_y,208,0,-3)
print(opt_k)
plt.figure(figsize=(12, 6))  

#'''
fig1 = plt.plot(range(208,0,-3), training_scores, color='red', linestyle='dashed', marker='o', markerfacecolor='red', markersize=5, label='train')
fig2 = plt.plot(range(208,0,-3), testing_scores, color='blue', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=5, label='test')
#'''
plt.legend(loc='lower left')
```

# iii. Since the computation time depends on the size of the training set, one may only use a subset of the training set. Plot the best error rate, which is obtained by some value of k, against the size of training set, when the size of training set is N  {10,20,30,...,210}.2 Note: for each N, select your training set by choosing the first N/3 rows of Class 0 and the first (N-N/3) rows of Class 1 in the training set you creatd in 1(b)iii. Also, for each N, select the optimal k from a set starting from k = 1, increasing by 5. For example, if N = 200, the optimal k is selected from {1,6,11,...,196}. This plot is called a Learning Curve.

```python
#score
def build_LearningCurve(df_NO,df_AB,start,end,step):
    score_=[]
    n=list(range(start,end,step))
    for N in range(start,end,step):
        train_x_=df_NO.iloc[:int(N/3),:-1].append(df_AB.iloc[:int(N-N/3),:-1])
        train_y_=df_NO.iloc[:int(N/3),6].append(df_AB.iloc[:int(N-N/3),6])
        test_x_=df_NO.iloc[int(N/3):,:-1].append(df_AB.iloc[int(N-N/3):,:-1])
        test_y_=df_NO.iloc[int(N/3):,6].append(df_AB.iloc[int(N-N/3):,6])
        training_scores,testing_scores,opt_k = find_ks(train_x_,train_y_,test_x_,test_y_,1,N,5)
        score_.append(1-abs(max(testing_scores)))
    maxscore_=abs(max(score_))
    opt_N=n[score_.index(maxscore_)]
    return(score_,opt_N)
    #plot
    #'''
score_,opt_N=build_LearningCurve(df_NO,df_AB,10,211,10)
plt.figure(figsize=(16, 6))  
plt.plot(range(10,211,10), score_, color='red', linestyle='dashed', marker='o',  
         markerfacecolor='blue', markersize=10)
plt.title('Learning Curve')  
plt.xlabel('N')  
plt.ylabel('Test Error')
    #'''
print(opt_N)
print(score_)
```

```python
#Manhattan Distance:
def find_k_D(train_x,train_y,test_x,test_y,start,end,step,p_,distance):
    score =[]
    k=list(range(start,end,step))
    for i in range(start,end,step):
        cl_ = KNeighborsClassifier(n_neighbors=i,p=p_,metric=distance)  
        cl_.fit(train_x,train_y)
        pred_i=cl_.predict(test_x)
        score.append(cl_.score(test_x,test_y))
        opt_k=k[score.index(max(score))]
    return(score,opt_k)

score, opt_k = find_k_D(train_x,train_y,test_x,test_y,1,197,5,1,'minkowski')
print(opt_k)

test_table = np.append([np.arange(1,197,5)],[np.array(score)],axis=0)
test_table = pd.DataFrame(test_table.T,columns=['k','test error'])
print(test_table)
plt.figure(figsize=(12, 6))  
plt.plot(range(1,197,5), score, color='red', linestyle='dashed', marker='o',  
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')  
plt.xlabel('K Value')  
plt.ylabel('score')
```

```python
#Mahalanobis Distance   
def MD(train_x,train_y,test_x,test_y):
    score =[]
    k=list(range(1,197,5))
    for i in range(1,197,5):
        cl_ = KNeighborsClassifier(n_neighbors=i, algorithm='brute', 
                               metric='mahalanobis', 
                               metric_params={'V': np.cov(train_x)})
        cl_.fit(train_x,train_y)
        print(cl_.score(test_x,test_y))
        score.append(cl_.score(test_x,test_y))
        opt_k=k[score.index(max(score))]
    return(score,opt_k)

score, opt_k = MD(train_x.values,train_y.values,test_x.values,test_y.values)
print('opt_k:',opt_k)
test_table = np.append([np.arange(1,197,5)],[np.array(score)],axis=0)
test_table = pd.DataFrame(test_table.T,columns=['k','test error'])
print(test_table)
plt.figure(figsize=(12, 6))  
plt.plot(range(1,197,5), score, color='red', linestyle='dashed', marker='o',  
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')  
plt.xlabel('K Value')  
plt.ylabel('score')  
```

---The majority polling decision can be replaced by weighted decision, in which the weight of each point in voting is proportional to its distance from the query/test data point. In this case, closer neighbors of a query point will have a greater influence than neighbors which are further away. Use weighted voting with Eu- clidean, Manhattan, and Chebyshev distances and report the best test errors when k  {1,6,11,16,...,196}.

```python
#add weights ,Eu
def find_k_D5(train_x,train_y,test_x,test_y,start,end,step):
    score =[]
    k=list(range(start,end,step))
    for i in range(start,end,step):
        cl_ = KNeighborsClassifier(n_neighbors=i,weights='distance')  
        cl_.fit(train_x,train_y)
        pred_i=cl_.predict(test_x)
        score.append(cl_.score(test_x,test_y))
        opt_k=k[score.index(max(score))]
    return(score,opt_k)
score,opt_k = find_k_D5(train_x,train_y,test_x,test_y,1,196,5)
print(opt_k)
print(score)
plt.figure(figsize=(12, 6))  
plt.plot(range(1,196,5), score, color='red', linestyle='dashed', marker='o',  
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')  
plt.xlabel('K Value')  
plt.ylabel('score')  
```
